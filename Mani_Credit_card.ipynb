```python
X=df.drop(['Time' , 'Class'], axis=1)
# Remove Time (just a sequence, not predictive)
# Remove Class (that's the target, not a feature)
# X now has only V1-V28 + Amount
```

```python
# Target class
y=df['Class'] # y is the fraud label (0 or 1)
```

```python
# V1 is normalized. Amount is NOT.
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler() 
#Create a StandardScaler object. This will normalize data using: (x - mean) / std
```

```python
X['Amount'] = scaler.fit_transform(X[['Amount']])
# fit_transform learns the mean and std of Amount, then normalizes it
# Amount now has mean ≈ 0 and std ≈ 1 (same as V1-V28)
```

```python
print(X['Amount'].describe())
```

```python
# Now Amount is on the same scale as V1-V28.
```

```python
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
print(f"Training set shape: {X_train.shape}")
print(f"Training labels: {y_train.shape}")
```

```python
print(f"Test set shape: {X_test.shape}")
print(f"Test labels: {y_test.shape}")
```

```python
# Check fraud % in training set
print(f"Training fraud %: {100 * y_train.mean():.2f}%")
```

```python
#  Check fraud % in test set

print(f"Test fraud %: {100 * y_test.mean():.2f}%")
```

```python
print(f"Training: {(y_train==1).sum()} frauds, {(y_train==0).sum()} legitimate")
print(f"Test: {(y_test==1).sum()} frauds, {(y_test==0).sum()} legitimate")
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42, k_neighbors=5)
# Create a SMOTE resampler. k_neighbors=5 means it looks at 5 nearest fraud neighbors
# to create new fake frauds.
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
print(f"\nAfter SMOTE:")
print(f"X_train_smote shape: {X_train_smote.shape}")
print(f"y_train_smote: {(y_train_smote==1).sum()} frauds, {(y_train_smote==0).sum()} legitimate")
```

```python
print(f"New ratio: {100 * y_train_smote.mean():.1f}% fraud (balanced!)")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000, random_state=42)
```

```python
lr.fit(X_train_smote, y_train_smote)
```

```python
y_pred_lr=lr.predict(X_test)
```

```python
y_pred_proba_lr = lr.predict_proba(X_test)[:, 1]
# [:, 1] = take the second column (probability of fraud)
```

```python
#Import evaluation metrics
from sklearn.metrics import confusion_matrix, roc_auc_score, f1_score
cm_lr = confusion_matrix(y_test, y_pred_lr)
tn_lr, fp_lr, fn_lr, tp_lr = cm_lr.ravel()

```

```python
# Measures how well model separates fraud from legitimate (0 to 1, higher is better)
roc_auc_lr = roc_auc_score(y_test, y_pred_proba_lr)
```

```python
f1_lr = f1_score(y_test, y_pred_lr)
# Harmonic mean of precision & recall (balances both)
```

```python
print(f"LOGISTIC REGRESSION RESULTS")
print(f"=" * 60)
print(f"\nConfusion Matrix:")
print(f"           Predicted")
print(f"           Legitimate  Fraud")
print(f"Actual Legitimate  {tn_lr:>7}  {fp_lr:>5}")
print(f"       Fraud       {fn_lr:>7}  {tp_lr:>5}")
```

```python
accuracy_lr = (tp_lr + tn_lr) / (tn_lr + fp_lr + fn_lr + tp_lr)
precision_lr = tp_lr / (tp_lr + fp_lr)
recall_lr = tp_lr / (tp_lr + fn_lr)

print(f"\nKey Metrics:")
print(f"Accuracy:  {100 * accuracy_lr:.2f}%")
print(f"Precision: {100 * precision_lr:.2f}%")
print(f"Recall:    {100 * recall_lr:.2f}%")
print(f"F1-Score:  {f1_lr:.4f}")
print(f"ROC-AUC:   {roc_auc_lr:.4f}")
```

```python
Catch 87.84% of frauds 
Cost: 120 false alarms per 8 real frauds 
```

```python
import xgboost as xgb
```

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f"Class weight: {scale_pos_weight:.2f}")
# 1 fraud is worth 579 legitimate transactions for learning")
```

```python
xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    verbosity=0
)
```

```python
n_estimators=100 = train 100 decision trees (boosting)
max_depth=5 = each tree is shallow (prevents overfitting)
learning_rate=0.1 = take small steps when learning
scale_pos_weight = weight fraud class higher
verbosity=0 = silent (no training messages)
```

```python
xgb_model.fit(X_train, y_train)
# Train on original imbalanced training data (NOT SMOTE).
# Why? XGBoost handles imbalance natively via scale_pos_weight. SMOTE is optional for it.
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```

```python
y_pred_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]
```

```python
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
tn_xgb, fp_xgb, fn_xgb, tp_xgb = cm_xgb.ravel()
```

```python
roc_auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)
```

```python
f1_xgb = f1_score(y_test, y_pred_xgb)
```

```python
print(f"\nXGBOOST RESULTS")
print(f"=" * 60)
print(f"\nConfusion Matrix:")
print(f"           Predicted")
print(f"           Legitimate  Fraud")
print(f"Actual Legitimate  {tn_xgb:>7}  {fp_xgb:>5}")
print(f"       Fraud       {fn_xgb:>7}  {tp_xgb:>5}")
```

```python
accuracy_xgb = (tp_xgb + tn_xgb) / (tn_xgb + fp_xgb + fn_xgb + tp_xgb)
precision_xgb = tp_xgb / (tp_xgb + fp_xgb)
recall_xgb = tp_xgb / (tp_xgb + fn_xgb)

print(f"\nKey Metrics:")
print(f"Accuracy:  {100 * accuracy_xgb:.2f}%")
print(f"Precision: {100 * precision_xgb:.2f}%")
print(f"Recall:    {100 * recall_xgb:.2f}%")
print(f"F1-Score:  {f1_xgb:.4f}")
print(f"ROC-AUC:   {roc_auc_xgb:.4f}")
```

```python
Logistic Regression (Bad Precision)

Caught 130 frauds
False alarms: ~1950 legitimate transactions flagged 
For every 1 fraud caught, you annoy 15 customers

XGBoost (Good Precision)

Caught 120 frauds
False alarms: ~92 legitimate transactions flagged ✓
For every 1 fraud caught, you annoy less than 1 customer
```
